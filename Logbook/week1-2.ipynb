```python
# 1. Area of a triangle with base 8 and height 12
base = 8
height = 12
area_triangle = 0.5 * base * height
print("1. Area of triangle:", area_triangle)
```

```python
# 2. Cube of 5 plus square of 4
cube_square = 5**3 + 4**2
print("2. Cube of 5 plus square of 4:", cube_square)
```

```python
# 3. Absolute difference between 7³ and 3⁷
abs_diff = abs(7**3 - 3**7)
print("3. Absolute difference between 7^3 and 3^7:", abs_diff)
```

```python
# 4. Sum of digits in the number 4729
num = 4729
sum_digits = sum(int(digit) for digit in str(num))
print("4. Sum of digits in 4729:", sum_digits)
```

```python
# 5. Surname repeated 3 times separated by dash
surname = "Gachhadar"
repeated_surname = "-".join([surname]*3)
print("5. Surname repeated:", repeated_surname)
```

```python
# 6. Name right-justified within 20 spaces
name = "Smriti"
print("6. Name right-justified:", name.rjust(20))
```

```python
# 7. Square root of 625 using math module
import math  # Make sure the math module is imported
sqrt_val = math.sqrt(625)
print("7. Square root of 625:", sqrt_val)
```

```python
# 8. 200 divided by 12 (quotient and remainder)
quotient = 200 // 12
remainder = 200 % 12
print("8. 200 divided by 12 -> Quotient:", quotient, ", Remainder:", remainder)
```

```python
# 9. Integer part of 9.9999 and rounded value to 2 decimal places
integer_part = int(9.9999)
rounded_val = round(9.9999, 2)
print("9. Integer part:", integer_part, ", Rounded to 2 decimals:", rounded_val)
```

```python
# 10. Unicode characters for codes: 80, 121, 116, 104, 111, 110
codes = [80, 121, 116, 104, 111, 110]
chars = "".join(chr(code) for code in codes)
print("10. Unicode characters:", chars)
```

```python
# Input from user
num = float(input("Enter a number: "))

# Check condition
if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```
