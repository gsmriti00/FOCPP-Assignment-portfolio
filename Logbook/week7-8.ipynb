```python
print("\nQUESTION 1:")
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    result = num1 / num2
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid input. Please enter numeric values.")
```

```python
def safe_int_input(prompt="Enter an integer: "):
    while True:
        try:
            value = int(input(prompt))
            return value
        except ValueError:
            print("Invalid input. Please enter a valid integer.")

print("\nQUESTION 2:")
number = safe_int_input()
print(f"You entered: {number}")
```

```python
print("\nQUESTION 3:")
filename = input("Enter filename to open: ")
try:
    with open(filename, "r") as f:
        content = f.read()
        print(f"File content:\n{content}")
except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
```

```python
def factorial(n):
    import math
    return math.factorial(n)

def square(n):
    return n ** 2

def cube(n):
    return n ** 3

print("\nQUESTION 4:")
print("Factorial of 5:", factorial(5))
print("Square of 4:", square(4))
print("Cube of 3:", cube(3))
```

```python
def km_to_miles(km):
    return km * 0.621371

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

print("\nQUESTION 5:")
print("10 km in miles:", km_to_miles(10))
print("25°C in Fahrenheit:", celsius_to_fahrenheit(25))
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2 * (self.length + self.width)

print("\nQUESTION 6:")
r1 = Rectangle(5, 3)
r2 = Rectangle(7, 4)
print(f"Rectangle 1 - Area: {r1.area()}, Perimeter: {r1.perimeter()}")
print(f"Rectangle 2 - Area: {r2.area()}, Perimeter: {r2.perimeter()}")
```

```python
class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
    def withdraw(self, amount):
        try:
            if amount > self.balance:
                raise ValueError("Insufficient balance.")
            self.balance -= amount
        except ValueError as e:
            print("Error:", e)
    def get_balance(self):
        return self.balance

print("\nQUESTION 7:")
account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
account.withdraw(2000)  # Should raise error
print("Balance:", account.get_balance())
```

```python
print("\nQUESTION 8:")
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
filtered = list(filter(lambda x: x > 10, squared))
print("Squared numbers:", squared)
print("Filtered (>10):", filtered)
```

```python
print("\nQUESTION 10:")
try:
    nums = list(map(int, input("Enter numbers separated by space: ").split()))
    if len(nums) < 2:
        raise ValueError("Need at least two numbers.")
    unique_nums = sorted(set(nums), reverse=True)
    second_largest = unique_nums[1]
    print("Second largest number:", second_largest)
except ValueError as e:
    print("Error:", e)
except IndexError:
    print("Error: Not enough unique numbers.")
```
