```python
import os

filename_q1 = "sample.txt"

# Create sample.txt if it doesn't exist
if not os.path.exists(filename_q1):
    sample_text = """Hello World! 123
Python programming is fun.
This is a test file."""
    with open(filename_q1, "w") as f:
        f.write(sample_text)
    print(f"'{filename_q1}' created with sample text.")

# Initialize counters
vowels_count = 0
consonants_count = 0
digits_count = 0
special_count = 0
vowels = "AEIOUaeiou"

# Open file and count characters
with open(filename_q1, "r") as file:
    for line in file:
        for ch in line:
            if ch.isdigit():
                digits_count += 1
            elif ch.isalpha():
                if ch in vowels:
                    vowels_count += 1
                else:
                    consonants_count += 1
            elif not ch.isspace():
                special_count += 1

# Display results
print("\nQUESTION 1 RESULTS:")
print(f"Vowels: {vowels_count}")
print(f"Consonants: {consonants_count}")
print(f"Digits: {digits_count}")
print(f"Special Characters: {special_count}")
```

```python

file1 = "file1.txt"
file2 = "file2.txt"
file3 = "merged.txt"

# Create file1 and file2 with sample content
for fname, text in [(file1, "This is file 1.\nLine two of file 1."),
                    (file2, "This is file 2.\nSecond line of file 2.")]:
    if not os.path.exists(fname):
        with open(fname, "w") as f:
            f.write(text)
        print(f"'{fname}' created with sample text.")

# Merge files
with open(file3, "w") as outfile:
    with open(file1, "r") as f1:
        outfile.write(f1.read())
    outfile.write("\n----- End of File 1 -----\n")
    with open(file2, "r") as f2:
        outfile.write(f2.read())

print(f"\nQUESTION 2: Files merged into '{file3}' successfully.")
```

```python
search_word = "file"
line_numbers = []

with open(file3, "r") as f:
    for idx, line in enumerate(f, 1):
        if search_word in line:
            line_numbers.append(idx)

print("\nQUESTION 3 RESULTS:")
print(f"The word '{search_word}' appears in line(s): {line_numbers}")
```

```python
import csv

students_csv = "students.csv"
# Create CSV if it does not exist
if not os.path.exists(students_csv):
    with open(students_csv, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Name", "Mark"])
        writer.writerow(["Anish", 88])
        writer.writerow(["Priya", 92])
        writer.writerow(["Sujan", 79])

# Read CSV and display formatted table with grades
print("\nQUESTION 4 RESULTS:")
print("{:<10} {:<5} {:<5}".format("Name", "Mark", "Grade"))
with open(students_csv, "r") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        mark = int(row["Mark"])
        if mark >= 90:
            grade = "A"
        elif mark >= 75:
            grade = "B"
        else:
            grade = "C"
        print(f"{row['Name']:<10} {mark:<5} {grade:<5}")
```

```python
from collections import Counter

filename_q5 = filename_q1  # reuse sample.txt

with open(filename_q5, "r") as f:
    words = f.read().split()
counter = Counter(words)
most_common_word, freq = counter.most_common(1)[0]

print("\nQUESTION 5 RESULTS:")
print(f"Most frequent word: '{most_common_word}', Occurrences: {freq}")
```

```python
unique_file = "unique_words.txt"
unique_words = sorted(set(words))
with open(unique_file, "w") as f:
    f.write("\n".join(unique_words))

print(f"\nQUESTION 6: Unique words written to '{unique_file}'.")
```

```python
vowel_lines_count = 0
with open(filename_q5, "r") as f:
    for line in f:
        if line and line[0].upper() in "AEIOU":
            vowel_lines_count += 1

print("\nQUESTION 7 RESULTS:")
print(f"Lines starting with a vowel: {vowel_lines_count}")
```

```python
reversed_file = "reversed.txt"
with open(filename_q5, "r") as f, open(reversed_file, "w") as rev:
    for line in f:
        rev.write(line[::-1] + "\n")

print(f"\nQUESTION 8: Reversed lines written to '{reversed_file}'.")
```

```python
employees_csv = "employees.csv"
# Sample employee CSV
if not os.path.exists(employees_csv):
    with open(employees_csv, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["Name", "Department", "Salary"])
        writer.writerow(["John", "HR", 5000])
        writer.writerow(["Jane", "IT", 6000])
        writer.writerow(["Jim", "HR", 7000])

dept_salaries = {}
with open(employees_csv, "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        dept = row["Department"]
        salary = float(row["Salary"])
        dept_salaries.setdefault(dept, []).append(salary)

print("\nQUESTION 9 RESULTS:")
for dept, sal_list in dept_salaries.items():
    avg_salary = sum(sal_list) / len(sal_list)
    print(f"{dept}: Average Salary = {avg_salary:.2f}")
```

```python
class Student:
    def __init__(self, id, name, address, admission_year, level, section):
        self.id = id
        self.name = name
        self.address = address
        self.admission_year = admission_year
        self.level = level
        self.section = section

    def display(self):
        print(f"ID: {self.id}, Name: {self.name}, Address: {self.address}, Year: {self.admission_year}, Level: {self.level}, Section: {self.section}")

print("\nQUESTION 10 RESULTS:")
# Example instance
s1 = Student(1, "Anish", "Kathmandu", 2025, "BSc", "A")
s1.display()
```

```python
class Employee:
    def __init__(self, empid, name, address, contact_number, spouse_name, num_children, salary):
        self.empid = empid
        self.name = name
        self.address = address
        self.contact_number = contact_number
        self.spouse_name = spouse_name
        self.num_children = num_children
        self.salary = salary

    def to_list(self):
        return [self.empid, self.name, self.address, self.contact_number, self.spouse_name, self.num_children, self.salary]

employees_list = [
    Employee(1, "John", "Kathmandu", "9800000000", "Mary", 2, 5000),
    Employee(2, "Jane", "Pokhara", "9800000001", "Paul", 1, 6000)
]

# Write to CSV with try/except
try:
    with open("employees_data.csv", "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["EmpID","Name","Address","Contact","Spouse","Children","Salary"])
        for emp in employees_list:
            writer.writerow(emp.to_list())
    print("\nQUESTION 11: Employees written to 'employees_data.csv'.")
except Exception as e:
    print(f"Error writing file: {e}")
```

```python
class Book:
    def __init__(self, title, author, available=True):
        self.title = title
        self.author = author
        self.available = available

    def issue(self):
        if self.available:
            self.available = False
            print(f"Book '{self.title}' issued successfully.")
        else:
            print(f"Book '{self.title}' is already issued.")

    def return_book(self):
        if not self.available:
            self.available = True
            print(f"Book '{self.title}' returned successfully.")
        else:
            print(f"Book '{self.title}' was not issued.")

# Example book list
books = [Book("Python 101", "John Doe"), Book("Learn Java", "Jane Smith")]

# Issue a book
books[0].issue()
# Return a book
books[0].return_book()
# Search for a book
search_title = "Learn Java"
found_books = [book for book in books if book.title == search_title]
print(f"\nQUESTION 12: Books found with title '{search_title}': {[book.title for book in found_books]}")
```
