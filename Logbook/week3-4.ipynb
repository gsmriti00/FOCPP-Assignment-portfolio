```python
def analyze_string(s):
    digits = sum(c.isdigit() for c in s)
    alphabets = sum(c.isalpha() for c in s)
    specials = len(s) - digits - alphabets
    return digits, alphabets, specials

# Example usage
string_input = "Hello123!@#"
d, a, s = analyze_string(string_input)
print("Digits:", d, "Alphabets:", a, "Special characters:", s)
```

```python
def is_perfect(n):
    divisors = [i for i in range(1, n) if n % i == 0]
    return sum(divisors) == n

# Example usage
num = 28
print(num, "is perfect:", is_perfect(num))
```

```python
def is_palindrome(num):
    return str(num) == str(num)[::-1]

# Example usage
num = 121
print(num, "is palindrome:", is_palindrome(num))
```

```python
def even_numbers(numbers):
    return sorted([n for n in numbers if n % 2 == 0])

# Example usage
nums = [5, 2, 9, 8, 4, 7]
print("Even numbers:", even_numbers(nums))
```

```python
import math

def square_root(x):
    return math.sqrt(x)

def absolute_value(x):
    return abs(x)

def floor_division(a, b):
    return a // b

def percentage(a, b):
    return (a / b) * 100

def average(a, b):
    return (a + b) / 2

# Example usage
print("Square root of 16:", square_root(16))
print("Absolute value of -5:", absolute_value(-5))
print("Floor division 10 // 3:", floor_division(10, 3))
print("Percentage 25 of 200:", percentage(25, 200))
print("Average of 10 and 20:", average(10, 20))
```
